## FACEBOOK API

```python
import requests
```

```python
id = 67563683055
url = 'http://graph.facebook.com/{}/picture?type=large'.format(id)
print(url)
```

```python
r = requests.get(url)
```

```python
type(r)
```

```python
r.content
```

```python
with open("{}_profile.jpg".format(id),"wb") as file:
    file.write(r.content)
```

## GOOGLE API

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json'
```

```python
parameters = {
    'address' : 'Santa Clara County'
}
```

```python
r = requests.get(url,params=parameters)
```

```python
r.url
```

```python
r.content
```

```python
r.content.decode('UTF-8')
```
