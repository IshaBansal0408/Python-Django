## Web Scraping

```python
import bs4
import requests
```

```python
url = 'https://www.passiton.com/inspirational-quotes'
```

```python
r = requests.get(url)
```

```python
r.content
```

```python
soup = bs4.BeautifulSoup(r.content)
```

```python
article_element = soup.findAll('div', {"class" : "col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top"})
```

```python
len(article_element)
```

```python
article_element[0].img.attrs['src']
```

```python
with open('inspiration.jpg','wb') as file:
    img_url = article_element[0].img.attrs['src']
    img_res = requests.get(img_url)
    file.write(img_res.content)
```

```python
for i,div in enumerate(article_element):
    with open('AllQuotes/inspiration_{}.jpg'.format(i),'wb') as file:
        try:
            img_url = div.img.attrs['src']
            print(img_url)
            img_res = requests.get(img_url)
            file.write(img_res.content)
        except:
            pass
```
